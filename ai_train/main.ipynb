# Stage 1: Manual Fuzzy Inference System (Sugeno)

Study Case: Bread Production Prediction based on Stock and Demand.

In this stage, we build a fuzzy system using human intuition to define Membership Functions (MF) and Rules manually.

```python
from helper import fuzzify, evaluate_rule, sugeno_defuzzification
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Membership Function (MF) Definitions

We define two inputs:
- **Stock**: Low, Medium, High.
- **Demand**: Low, Medium, High.

And one output:
- **Production**: Constants (Zero-order Sugeno).

```python
# Membership Function configurations for Stock (0-100)
stock_mf = {
    'Low': ('left_shoulder', [20, 40]),
    'Medium': ('triangular', [30, 50, 70]),
    'High': ('right_shoulder', [60, 80])
}

# Membership Function configurations for Demand (0-200)
demand_mf = {
    'Low': ('left_shoulder', [50, 100]),
    'Medium': ('triangular', [80, 120, 160]),
    'High': ('right_shoulder', [140, 180])
}
```

## 2. Rule Base Definitions

Based on human intuition:

```python
rule_base = [
    {'if': [('demand', 'High'), ('stock', 'Low')], 'then': 250},   # Very High
    {'if': [('demand', 'High'), ('stock', 'Medium')], 'then': 200}, # High
    {'if': [('demand', 'High'), ('stock', 'High')], 'then': 125},   # Medium
    {'if': [('demand', 'Medium'), ('stock', 'Low')], 'then': 200}, # High
    {'if': [('demand', 'Medium'), ('stock', 'Medium')], 'then': 125}, # Medium
    {'if': [('demand', 'Medium'), ('stock', 'High')], 'then': 50},   # Low
    {'if': [('demand', 'Low'), ('stock', 'Low')], 'then': 50},    # Low
    {'if': [('demand', 'Low'), ('stock', 'Medium')], 'then': 50}, # Low
    {'if': [('demand', 'Low'), ('stock', 'High')], 'then': 0}      # Stop
]
```

## 3. Inference Function

```python
def predict_production(stock_val, demand_val):
    # Fuzzification
    s_fuzzy = fuzzify(stock_val, stock_mf)
    d_fuzzy = fuzzify(demand_val, demand_mf)
    
    memberships = {'stock': s_fuzzy, 'demand': d_fuzzy}
    
    # Rule Evaluation
    firing_strengths = []
    rule_outputs = []
    
    for rule in rule_base:
        w, z = evaluate_rule(memberships, rule)
        firing_strengths.append(w)
        rule_outputs.append(z)
        
    # Defuzzification (Sugeno)
    return sugeno_defuzzification(firing_strengths, rule_outputs)

# Test Case 1: Low Stock (10), High Demand (190)
print(f"Predicted Production (Stock 10, Demand 190): {predict_production(10, 190)}")

# Test Case 2: High Stock (90), Low Demand (20)
print(f"Predicted Production (Stock 90, Demand 20): {predict_production(90, 20)}")

# Test Case 3: Medium Conditions (50, 120)
print(f"Predicted Production (Stock 50, Demand 120): {predict_production(50, 120)}")
```
